```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wine_ratings = pd.read_csv('wine_ratings.csv')
vintage_ratings = pd.read_csv('vintage_ratings.csv')
print(wine_ratings.head())
print(vintage_ratings.head())
```

```python
%matplotlib inline
plt.figure(figsize=(20, 6))
plt.title('Average Rating of Wines per Country')
plt.ylabel('Average Rating')
plt.bar(wine_ratings['name'], wine_ratings['average_rating'], color='#a32e61')

plt.show()
```

```python
%matplotlib inline
plt.figure(figsize=(20, 6))
plt.title('Average Rating of Vintage Wines per Country')
plt.ylabel('Average Rating')
plt.bar(vintage_ratings['name'], vintage_ratings['average_rating'], color='#a32e61')
plt.show()
```
